```python
#pertanyaan
Pertanyaan = {
    1 : {
        "pertanyaan" : "Suhu tubuh manusia normal 37-40 derajat Celcius",
        "jawaban_benar" : "0",
        "jawaban_salah" : "1",
        "komentar_benar" : "Normal juga kamu yaa, xixixi",
        "komentar_salah" : "Xixixi, sepertinya kamu demam setiap hari!"
    },
    2 : {
        "pertanyaan" : "Obat tablet lambung berwarna hijau",
        "jawaban_benar" : "1",
        "jawaban_salah" : "0",
        "komentar_benar" : "Aku curiga kamu punya sakit lambung, xixixi",
        "komentar_salah" : "Kau kira warnanya apa?! Xixixi"
    },
    3 : {
        "pertanyaan" : "Mandi malam menyebabkan Reumatik",
        "jawaban_benar" : "0",
        "jawaban_salah" : "1",
        "komentar_benar" : "Xixixi, aku selalu mandi malam dan tidak Reumatik!",
        "komentar_salah" : "Kamu terlalu percaya mitos, xixixi"
    },   
}

#fungsi cek jawaban
def cekJawaban(pertanyaan, jawaban, kesempatan):
  if Pertanyaan[pertanyaan]['jawaban_benar'] == jawaban:
    print (Pertanyaan[pertanyaan]['komentar_benar'])
    print (f"Kamu masih memiliki {kesempatan} kesempatan") 
    return True
  elif Pertanyaan[pertanyaan]['jawaban_salah'] == jawaban:
    print (Pertanyaan[pertanyaan]['komentar_salah'])
    print ("Kesempatan kamu menjadi " + str(kesempatan-1))
    return False
  else:
    print(f"Jawab dengan 1 atau 0, sekarang kesempatanmu tinggal {kesempatan-1}")
    return False

# Intro 
def awalan():
  print("""Selamat datang di Jembatan Pancing
  Jawab pertanyaan dariku untuk bisa melewati jembatan ini
  Gunakan 1 untuk jawaban Iya dan 0 untuk jawaban Tidak
  Kesempatanmu menjawab hanya 3 kali
  """)
  input("Tekan apa saja untuk memulai: ")
  print ("Mari kita mulai, xixixi!")
  return True;

# Penutupan
def penutup():
  if kesempatan > 0:
    print ("\nSekarang kamu bisa melewati Jembatan Pancing ini, kuharap kita tidak bertemu lagi!")
  else:
    print ("\nXixixi, pergi sana! Jangan berharap dapat melewati Jembatan Pancing ini!")

# Logika Kuis
intro = awalan()

while True:
  kesempatan = 3
  print(f"Kamu punya {kesempatan} kesempatan")
  for pertanyaan in Pertanyaan:
    while kesempatan > 0:
      print("\n")
      print(Pertanyaan[pertanyaan]['pertanyaan'])
      jawaban = input("Iya atau Tidak : ")
            
      # cek Jawaban jika bener
      cek_Jawaban = cekJawaban(pertanyaan, jawaban, kesempatan)
      if cek_Jawaban:
        print("\nLanjuuuuuuuuuut")
        break
      # jika kondisi jawaban salah atau tidak menjalankan kondisi if diatas maka jalankan else
      kesempatan-=1
  break

  # break berfungsi untuk keluar perulangan

outro = penutup()
```
